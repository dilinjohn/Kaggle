```python
train = pd.read_csv('/kaggle/input/nlp-getting-started/train.csv')
test = pd.read_csv('/kaggle/input/nlp-getting-started/test.csv')
```

```python
train.head()
```

```python
display(train.isnull().sum())
print(train.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import cross_val_score
```

```python
train['text']
```

```python
cvect = CountVectorizer(stop_words='english')

# looking at the first 5 tweets
example_train = cvect.fit_transform(train['text'][0:5])
example_train
```

```python
display(example_train[0].todense())
display(example_train[0].toarray())
```

```python
# on the whole train
import time
start=time.time()
train_csr = cvect.fit_transform(train['text'])
print('time taken:', time.time()-start)
```

```python
# apply the transform method on text using tokens in train or the vocabulary from the train
# use only the transform and not the fit method on test
test_csr = cvect.transform(test['text'])
```

### Since this is a high dimensional vector, we will apply some dimension reduction techniques
- RidgeClassifier

```python
ridge = RidgeClassifier()
X = train_csr
y = train['target']

# scoring metric is : F1
f1_scores = cross_val_score(ridge, X, y, cv=5, scoring='f1')
f1_scores
```

```python
ridge.fit(X, y)
```

```python
predictions = ridge.predict(test_csr)
```

```python
submission = pd.read_csv('/kaggle/input/nlp-getting-started/sample_submission.csv')
submission['target'] = predictions
submission
```
